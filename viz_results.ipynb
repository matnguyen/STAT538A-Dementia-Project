```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
labels = ['LR', 'MELR-S', 'MELR-V', 'MELR-SV', 'RF', 'SVM']
full_model = [0.792, 0.581, 0.740, 0.581, 0.816, 0.581]
select_model = [0.792, 0.754, 0.768, 0.765, 0.813, 0.813]

x = np.arange(len(labels))
width=0.35

fig, ax = plt.subplots(figsize=(10,8))
rects1 = ax.bar(x - width/2, full_model, width, label='Full model')
rects2 = ax.bar(x + width/2, select_model, width, label='Selected model')

ax.set_ylabel('Accuracy')
ax.set_title('Accuracy for predicting `Demented` response')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 1)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
fig.savefig('figures/demented.png')

plt.show()
```
